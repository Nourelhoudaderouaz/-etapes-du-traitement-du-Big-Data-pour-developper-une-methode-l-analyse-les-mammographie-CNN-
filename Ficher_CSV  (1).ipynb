```python
import pandas as pd
import os

# Chemin du fichier CSV de sortie
nouveau_fichier_path = '/content/drive/MyDrive/Matlab/NV.csv'

# Liste pour stocker les annotations des images trouvées
annotations_trouvees = []
nb_photos_trouvees = 0
nb_photos_non_trouvees = 0

# Parcourir les lignes du DataFrame df_train
for index, row in df_train.iterrows():
    image_path = row['path']
    
    # Chercher l'image dans le dossier spécifié
    if os.path.exists(image_path):
        annotations_trouvees.append(row)
        print("J'ai trouvé l'image.")
        nb_photos_trouvees += 1 
    else:
        nb_photos_non_trouvees += 1
        print(f"L'image {image_path} n'a pas été trouvée.")

# Vérifier si des annotations ont été trouvées
if annotations_trouvees:
    # Créer un nouveau DataFrame avec les annotations trouvées
    df_resultats = pd.DataFrame(annotations_trouvees, columns=df_train.columns)
    # Enregistrer le DataFrame des résultats dans un nouveau fichier CSV
    df_resultats.to_csv(nouveau_fichier_path, index=False)
    print(f"Les annotations ont été enregistrées dans le nouveau fichier {nouveau_fichier_path}.")
else:
    print("Aucune annotation trouvée.")

# Afficher le nombre de photos trouvées et non trouvées
print(f"Nombre de photos trouvées : {nb_photos_trouvees}")
print(f"Nombre de photos non trouvées : {nb_photos_non_trouvees}")
```
